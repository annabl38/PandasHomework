# PyCity Schools Analysis



<h1>Setup & Import data </h1>

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os 

# File to Load (Remember to Change These)
school_data_to_load = os.path.join("Resources", "schools_complete.csv")
student_data_to_load = os.path.join("Resources", "students_complete.csv")

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

<h1>District Summary</h1>


```python
school_count=school_data["school_name"].count()
student_count=school_data_complete["student_name"].count()
total_budget=school_data["budget"].sum()
avg_math=student_data["math_score"].mean()
avg_read=student_data["reading_score"].mean()
math_pass_count=school_data_complete[school_data_complete["math_score"]>=70].count()['math_score']
math_pass=math_pass_count/student_count
read_pass_count=student_data[student_data["reading_score"]>=70].count()['reading_score']
read_pass=read_pass_count/student_count
#student_data["overall_score"]=(student_data["math_score"]+student_data['reading_score'])/2
#over_pass_count=student_data[student_data["overall_score"]>=70].count()['overall_score']
over_pass=(math_pass+read_pass)/2

district_df=pd.DataFrame({"Total Schools":school_count,
         "Total Students":student_count,
         "Total Budget":total_budget,
         "Average Math Score":avg_math,
         "Average Reading Score":avg_read,
         "% Passing Math": math_pass,
         "% Passing Reading": read_pass,
         "% Overall Passing Rate": over_pass}, index=[0])
district_df["Total Budget"]=district_df["Total Budget"].map("${:,.2f}".format)
district_df["Average Math Score"]=district_df["Average Math Score"].map("{:.2f}".format)
district_df["Average Reading Score"]=district_df["Average Reading Score"].map("{:.2f}".format)
district_df["% Passing Math"]=district_df["% Passing Math"].map("{:,.2%}".format)
district_df["% Passing Reading"]=district_df["% Passing Reading"].map("{:,.2%}".format)
district_df["% Overall Passing Rate"]=district_df["% Overall Passing Rate"].map("{:,.2%}".format)

district_df
```

<h1> School Summary </h1>

```python
group_schools=school_data_complete.groupby(['school_name'])

school_students=group_schools['student_name'].count()
school_budget=group_schools['budget'].first()
school_type=group_schools['type'].first()
budget_per=school_budget/school_students

school_avg_math=group_schools['math_score'].mean()
school_avg_read=group_schools['reading_score'].mean()

all_math_pass=school_data_complete.loc[school_data_complete['math_score']>=70,:]
all_math_counts=all_math_pass.groupby(['school_name']).count()['math_score']
per_pass_math=(all_math_counts/school_students)*100

all_read_pass=school_data_complete.loc[school_data_complete['reading_score']>=70,:]
all_read_counts=all_read_pass.groupby(['school_name']).count()['reading_score']
per_pass_read=(all_read_counts/school_students)*100

over_pass_all=(per_pass_math+per_pass_read)/2



school_sum_raw=pd.DataFrame({'School Type': school_type,
                         'Total Students': school_students,
                        'School Budget': school_budget,
                        'Budget per Student':budget_per,
                        'Average Math Score': school_avg_math,
                        'Average Reading Score': school_avg_read,
                        '% Passing Math': per_pass_math,
                        '% Passing Reading': per_pass_read,
                        'Overall Passing Rate': over_pass_all})
school_sum=school_sum_raw.copy()
school_sum["School Budget"]=school_sum["School Budget"].map("${:,.2f}".format)
school_sum["Budget per Student"]=school_sum["Budget per Student"].map("${:,.2f}".format)
school_sum["Average Math Score"]=school_sum["Average Math Score"].map("{:.2f}".format)
school_sum["Average Reading Score"]=school_sum["Average Reading Score"].map("{:.2f}".format)
school_sum["% Passing Math"]=school_sum["% Passing Math"].map("{:,.2f}".format)
school_sum["% Passing Reading"]=school_sum["% Passing Reading"].map("{:,.2f}".format)
school_sum['Overall Passing Rate']=school_sum["Overall Passing Rate"].map("{:,.2f}".format)
school_sum
```

## Top Performing Schools (By Passing Rate)

```python
top_schools=school_sum.sort_values(['Overall Passing Rate'],ascending=False)
top_schools.head()
```

## Bottom Performing Schools (By Passing Rate)

```python
bottom_schools=school_sum.sort_values(['Overall Passing Rate'],ascending=True)
bottom_schools.head()
```

## Math Scores by Grade

```python
ninth_math={}
tenth_math={}
eleventh_math={}
twelfth_math={}
for index, row in school_data_complete.iterrows():
    if row['grade']=='9th' and row['school_name'] not in  ninth_math:
        ninth_math[row['school_name']]=[row['math_score']]
    elif row['grade']=='9th':
        ninth_math[row['school_name']].append(row['math_score'])
    elif row['grade']=='10th' and row['school_name'] not in  tenth_math:
        tenth_math[row['school_name']]=[row['math_score']]
    elif row['grade']=='10th':
        tenth_math[row['school_name']].append(row['math_score'])
    elif row['grade']=='11th' and row['school_name'] not in  eleventh_math:
        eleventh_math[row['school_name']]=[row['math_score']]
    elif row['grade']=='11th':
        eleventh_math[row['school_name']].append(row['math_score'])    
    elif row['grade']=='12th' and row['school_name'] not in  twelfth_math:
        twelfth_math[row['school_name']]=[row['math_score']]
    elif row['grade']=='12th':
        twelfth_math[row['school_name']].append(row['math_score'])    
ninth_math_avg={}
tenth_math_avg={}
eleventh_math_avg={}
twelfth_math_avg={}
for key,value in ninth_math.items():
     ninth_math_avg[key]=(sum(value)/len(value))
for key,value in tenth_math.items():
     tenth_math_avg[key]=(sum(value)/len(value))
for key,value in eleventh_math.items():
     eleventh_math_avg[key]=(sum(value)/len(value))
for key,value in twelfth_math.items():
     twelfth_math_avg[key]=(sum(value)/len(value))
    
math_by_grade=pd.DataFrame({'9th':ninth_math_avg,
                           '10th':tenth_math_avg,
                           '11th':eleventh_math_avg,
                           '12th':twelfth_math_avg})
math_by_grade
```
